# Principal Component Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import urllib.request
```

```python
data = pd.read_table('ex7data1.txt', sep=',', header=None, dtype=np.float32)
data
```

```python
X_train = data.values
```

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(1, 1, 1)
ax.axis([0.5, 6.5, 2, 8])
ax.plot(X_train[:,0], X_train[:,1], marker='.', color='#0F00FF', markersize=10, linestyle='') 
```

```python
model = PCA(n_components=1)

model.fit(X_train)
```

```python
model.components_
```

```python
X_red = model.transform(X_train)
X_red
```

```python
X_pred = model.inverse_transform(X_red)
X_pred
```

```python
X_train
```

```python
np.mean(np.sum(np.square(X_train-X_pred), axis=1))
```

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(1, 1, 1)
ax.axis([0.5, 6.5, 2, 8])
ax.plot(X_train[:,0], X_train[:,1], marker='.', color='#0F00FF', markersize=10, linestyle='') 
ax.plot(X_pred[:,0], X_pred[:,1], marker='.', color='#FF00AE', markersize=10)
```

```python
m,n = X_train[:100,:].shape
a = np.ceil(np.sqrt(n)).astype(int)
b = np.ceil(np.sqrt(m)).astype(int)
fig, axes = plt.subplots(figsize=(10,10),nrows=b, ncols=b, sharex=True, sharey=True)
fig.subplots_adjust(wspace=-0.01, hspace=-0.01)
for i in range(b):
    for j in range(b):
        img = np.rot90(np.fliplr(X_train[i*b+j,:].reshape((a,a))))
        axes[i,j].set_yticklabels([])
        axes[i,j].set_xticklabels([])
        axes[i,j].imshow(img,cmap='gray')
```

```python
model = PCA(n_components=100)
model.fit(X_train)
```

```python
model.components_
```

```python
X_red = model.transform(X_train)
X_red
```

```python
X_pred = model.inverse_transform(X_red)
X_pred
```

```python
X_train
```

```python
np.mean(np.sum(np.square(X_train-X_pred), axis=1))
```
